```python
# classes
class Dog:
    """Represent a dog."""
    def __init__(self, name):
        """Initialize dog object."""
        self.name = name
    def sit(self):
        """Simulate sitting."""
        print(f"{self.name} is sitting.")

my_dog = Dog('Peso')
print(f"{my_dog.name} is a great dog!")
my_dog.sit()


## inheritence
class SRCHDog(Dog):
    """Represent a search dog."""
    def __init__(self, name):
        """Initialize the sardog."""
        super().__init__(name)
    def search(self):
        """Simulate searching."""
        print(f"{self.name} is searching.")
my_dog = SRCHDog('Willie')
print(f"{my_dog.name} is a search dog.")
my_dog.sit()
my_dog.search()
```
